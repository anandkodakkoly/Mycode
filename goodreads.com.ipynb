```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

def quoted_info(url):
    req = requests.get(url).text
    ip = BeautifulSoup(req , 'html.parser')
    df = pd.DataFrame()
    quotes = [ list(i.children)[0].strip()  for i in ip.find_all('div', class_='quoteText')]
    author_name = [ i.span.text.strip()  for i in ip.find_all('div', class_='quoteText') ]
    likes = [i.text.rstrip(' likes') for i in ip.find_all('a' ,title='View this quote' ) ]
    df['Authorname'] = author_name
    df['Quotes'] = quotes
    df['No of Likes'] = likes
    return df

df = pd.DataFrame()

url = 'https://www.goodreads.com/quotes?page={}'
from tqdm import tqdm
for i in tqdm(range(1,100)):
    details = quoted_info(url.format(i))
    df = pd.concat([df,details])
df
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

def quoted_info(url):
    req = requests.get(url).text
    ip = BeautifulSoup(req , 'html.parser')
    df = pd.DataFrame()
    quotes = [ list(i.children)[0].strip()  for i in ip.find_all('div', class_='quoteText')]
    author_name = [ i.span.text.strip()  for i in ip.find_all('div', class_='quoteText') ]
    likes = [i.text.rstrip(' likes') for i in ip.find_all('a' ,title='View this quote' ) ]
    df['Authorname'] = author_name
    df['Quotes'] = quotes
    df['No of Likes'] = likes
    return df

master_d = {}

url = 'https://www.goodreads.com/quotes?page={}'
from tqdm import tqdm
for i in tqdm(range(1,101)):
    master_d[i] = quoted_info(url.format(i))
```
